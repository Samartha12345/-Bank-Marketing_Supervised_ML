```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
sns.histplot(x='age',data=df,kde='True',palette='viridis',bins=20)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="education", hue="y", palette="pastel")
plt.title("Relationship Between Education and Target (y)")
plt.xlabel("Education Levels")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Subscribed (y)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='month', data=df,palette='viridis') 
plt.xlabel("Month in which person was contacted")
plt.ylabel("Count")
plt.title('Month Distribution', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='loan', hue='y', palette='mako')
plt.title('Distribution of loan with respect to y', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='y', loc='upper right');
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='job', hue='y', palette='coolwarm')
plt.title('Distribution of Job with respect to y', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='y', loc='upper right');
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="job", hue="y", palette="viridis")
plt.title("Relationship Between Job and Target (y)")
plt.xlabel("Jobs")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Subscribed (y)")
plt.tight_layout()
plt.show()
```
